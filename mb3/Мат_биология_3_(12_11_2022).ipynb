```python
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt
```

```python
def func_tiktok(x):
    while x > 1:
        y.append(x)
        if (x % 2) == 0 :
            # print(str(x) + ' это четное число')
            x = x // 2
        else:
            # print(str(x) + ' это нечетное число')
            x = 3*x + 1

    return
```

```python
x = int(input('enter x: '))
y = []

func_tiktok(x)
```

```python
len_arr = len(y)
x = np.linspace(0, len_arr, len_arr)

plt.figure(figsize=(20,10))
plt.plot(x,y, linewidth=1)

plt.grid()
plt.xlabel('x')
plt.ylabel('y')
plt.title('гипотеза коллатца')

plt.show()
```
