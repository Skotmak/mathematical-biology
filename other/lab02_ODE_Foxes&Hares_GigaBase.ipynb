# Жертвы и хищники — модель динамики биологических систем

Экосистемы – это природные ансамбли, в которых взаимодействуют друг с другом популяции живых и неживых организмов . Но какие законы управляют динамикой этих сложных систем? И, в частности, как виды хищников и жертв взаимодействуют друг с другом?

В 1925-1926 годах Альфред Лотка и Вито Вольтерра, два математика, изучали эту проблему и независимо друг от друга разработали модель дифференциального уравнения для решения в случае экосистемы, состоящей из двух видов животных, один из которых действует как вид хищника, а другой - как его жертва: модель «жертва-хищник».

# Модель Лотка-Вольтерра
Основная идея [модели Лотки-Вольтерры](https://ru.wikipedia.org/wiki/%D0%9C%D0%BE%D0%B4%D0%B5%D0%BB%D1%8C_%D0%9B%D0%BE%D1%82%D0%BA%D0%B8_%E2%80%94_%D0%92%D0%BE%D0%BB%D1%8C%D1%82%D0%B5%D1%80%D1%80%D1%8B) состоит в том, чтобы рассмотреть идеальный сценарий, в котором сосуществуют только два вида: хищники и их жертвы. Есть несколько основных предположений:

*   хищники могут питаться только популяцией жертв;
*   общее количество потребляемой хищниками пищи (т.е. количество съеденной жертвы) в единицу времени пропорционально числу встреч, происходящих между жертвами и хищниками;
*   количество встреч между жертвами и хищниками пропорционально размеру обеих популяций;
*   каждому хищнику требуется минимальное количество пищи в единицу времени, чтобы выжить и размножаться;
*   у каждой жертвы есть неиссякаемый источник пищи, позволяющий ей размножаться в отсутствие хищников.



Если запас пищи для жертвы неограничен, разумно предположить, что скорость роста этой популяции $ x $ будет пропорциональна ее текущему
размеру (с учетом большего количества потенциальных связей):

$ \frac{dx}{dt} = \alpha x(t)$

интегрируем и получаем

$ x(t) = x(0)e^{\alpha t} $

где $ \alpha > 0 $ — коэффициент роста, т. е. коэффициент размножения на жертвы.



> $ P(t) = P(0)e^{rt} $ — этот закон называется [Мальтузианская модель роста](https://ru.wikipedia.org/wiki/%D0%9C%D0%B0%D0%BB%D1%8C%D1%82%D1%83%D0%B7%D0%B8%D0%B0%D0%BD%D1%81%D0%BA%D0%B0%D1%8F_%D0%BC%D0%BE%D0%B4%D0%B5%D0%BB%D1%8C_%D1%80%D0%BE%D1%81%D1%82%D0%B0) и предсказывает экспоненциальный рост населения/популяции со временем. Он довольно хорошо описывает, что происходит с определенными бактериями или культурами клеток в течение короткого времени.


Без хищников популяция $ x $ неограниченно растет со временем. Но с хищниками оно уменьшится пропорционально количеству хищников $ y $, умноженному на количество жертв $ x $, из-за **количества взаимодействий между двумя видами**.

$ \frac{dx}{dt} = \alpha x(t) - \beta x(t) y(t)  $

где $ \beta $ — скорость, с которой происходят фатальные взаимодействия.

Для хищников мы должны предположить, что в отсутствие жертвы популяция $ y $ будет уменьшаться по мере того, как ее кормовая база иссякает:

$ \frac{dy}{dt} = - \gamma y(t) $

интегрируем и получаем

$ y(t) = y(0)e^{-\gamma t} $

где $ \gamma $ — коэффициент вымирания хищника.

Полученное предсказывает окончательное исчезновение хищников.


С другой стороны, хищники также будут расти пропорционально взаимодействию двух видов, то есть:

$ \frac{dy}{dt} = - \gamma y(t) + \delta x(t) y(t) $

где $ \delta $ — скорость взаимодействия между двумя видами.


Собирая все вместе, получаем систему дифференциальных уравнений, описывающих **скорость роста двух популяций** $ x $ (жертвы) и $ y $ (хищники):

$ 
\begin{equation*}
 \begin{cases}
   x'(t) = \alpha x(t) - \beta x(t) y(t)
   \\
   y'(t) = - \gamma y(t) + \delta x(t) y(t)
 \end{cases}
\end{equation*}
$

или убрав лишнее, чтобы проще было воспринимать

$ 
\begin{equation*}
 \begin{cases}
   \dot{x} = (\alpha  - \beta y)x
   \\
   \dot{y} = (- \gamma + \delta x) y
 \end{cases}
\end{equation*}
$

# Равновесие системы
Популяции достигают точки равновесия, когда обе поддерживают постоянный уровень с течением времени, т.е.:

$ x(t) = x_0 $

$ y(t) = y_0 $

В этом случае темпы роста, т.е. производные, равны нулю. Подстановкой получаем уравнения равновесия:

$ 
\begin{equation*}
 \begin{cases}
   0 = (\alpha  - \beta y)x
   \\
   0 = (- \gamma + \delta x) y
 \end{cases}
\end{equation*}
$

У нас есть два возможных решения:

$ (x_0, y_0) = (0, 0) $

$ (x_0, y_0) = (\frac{\gamma}{\delta}, \frac{\alpha}{\beta}) $

В первом случае оба вида вымерли.

Вместо этого второй случай соответствует ситуации, в которой число жертв точно соответствует количеству пищи, которое удерживает стационарную популяцию хищников. Подробнее об этом позже.

# Лисы и зайцы
Сделаем пример. Мы предполагаем, что вначале есть две популяции, 5 лис (хищники) и 10 зайцев (жертвы). После решения модели путем численного интегрирования мы анализируем временную [коэволюцию](https://ru.wikipedia.org/wiki/%D0%9A%D0%BE%D1%8D%D0%B2%D0%BE%D0%BB%D1%8E%D1%86%D0%B8%D1%8F) видов.

```python
import numpy as np
from scipy import integrate
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Параметры, т.е. коэффициенты

alpha = 1.
beta = 0.1
delta = 0.075
gamma = 1.5
```

```python
def dX_dt(x, t=0):
    """
    Возвращает скорость роста лис и зайцев.
    """

    result = np.array([
        (alpha - beta * x[1]) * x[0],
        (-gamma + delta * x[0]) * x[1] 
        ])
    return result
```

```python
# Точки равновесия
X_f0 = np.array([0. , 0.])
X_f1 = np.array([gamma / delta, alpha / beta])
```

```python
# Массив времени
t = np.linspace(0, 20, 1000)
```

```python
# Заданные условия (количество зайцев и лис)
X0 = np.array([10, 5])
```

```python
# Интегрируем систему.
X = integrate.odeint(dX_dt, X0, t)
hares, fox = X.T
```

```python
# График времени коэволюции
size_graph = (15, 7)
fig, ax = plt.subplots(figsize=size_graph)

plt.plot(t, hares, 'r-', label = 'Зайцы')
plt.plot(t, fox, 'b-', label = 'Лисы')
plt.grid(True)
plt.legend(loc='best')
plt.xlabel('Время (t)')
plt.ylabel('Популяции')
plt.title('Время коэволюции')

plt.show()
```

Модель имеет периодический характер. На самом деле, увеличение жертв стимулирует рост хищников. Однако при увеличении хищников популяция жертв уменьшается и достигает минимума, так что и хищники уменьшаются. Когда хищники уменьшаются, происходит новый прирост жертв и так далее. Таким образом, эта динамика приводит к непрерывному циклу роста и упадка.

Периодическое поведение также хорошо иллюстрируется фазовым графиком, параметризованным для разных начальных размеров популяций и изображенным на следующем графике. Тот факт, что кривые замкнуты, указывает на периодическую связь между двумя переменными.

```python
# Размеры графика
fig, ax = plt.subplots(figsize = size_graph)

# Масштабный коэффициент для начальных условий.
values  = np.linspace(0.2, 1, 5)

# Цвета для каждой траектории.
vcolors = plt.cm.autumn_r(np.linspace(0.2, 1., len(values)))

# Нанесем на график все траектории.
for value, col in zip(values, vcolors):
    # Решаем новую систему
    X0 = value * X_f1
    X = integrate.odeint(dX_dt, X0, t)

    # Строим график решения
    plt.plot( X[:,0], X[:,1], lw = 3.5 * value, color = col,
        label = f'X0=({X0[0]:.2f}, {X0[1]:.2f})')
    
# Определяем сетку и вычисляем направление для каждой точки.
n_points = 15
ymax = plt.ylim(ymin = 0)[1]
xmax = plt.xlim(xmin = 0)[1]
x = np.linspace(0, xmax, n_points)
y = np.linspace(0, ymax, n_points)
X1, Y1 = np.meshgrid(x, y)

# Темп роста для всей сети.
DX1, DY1 = dX_dt([X1, Y1])

# Нормализуем и стабилизируем.
M = np.hypot(DX1, DY1)
M[M == 0] = 1.
DX1 /= M
DY1 /= M

plt.title('Фазовый портрет')
plt.quiver(X1, Y1, DX1, DY1, M, pivot = 'mid', cmap = plt.cm.jet)
plt.xlabel('Зайцы')
plt.ylabel('Лисы')
plt.legend()
plt.grid()
plt.xlim(0, xmax)
plt.ylim(0, ymax)
plt.show()
```

# Анализ результатов
Зададим закономерные вопросы:

*   может ли один вид вымереть независимо от другого?
*   могут ли популяции стабилизироваться и достичь точки равновесия?

Ответ на первый вопрос исходит равновесных решений, которые нам говорят, что каждая популяция может быть нулевой только в том случае, если другая тоже равна нулю. Виды не могут вымирать в разное время, они обязательно должны вымирать одновременно. Но замкнутый вид траекторий свидетельствует о том, что в модели это условие никогда не встречается (в реальном мире больше факторов влияющих на популяции).

Также никогда не достигается стабилизация. Система циклически повторяется, тогда как решение, стремящееся к равновесию, будет иметь спиральную диаграмму, обращённую во внутрь.
